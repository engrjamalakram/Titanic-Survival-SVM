# Step 10.2 — Handling Missing Values in the Age Column

The **Age** feature is one of the most important variables in the Titanic dataset.

However, some passengers do not have their age recorded.

Since Support Vector Machine (SVM) cannot work with missing numerical values, we must replace these missing entries.

Instead of using the mean, we will use the **median**, which is less affected by extreme values (outliers) and provides a more robust estimate of the central value.

````python
# ================================================================
# STEP 10.2 : HANDLE MISSING AGE VALUES
# ================================================================

# Count missing Age values before filling.

print("="*60)
print("Missing Age Values Before Cleaning")
print("="*60)

print(df["Age"].isnull().sum())

# ------------------------------------------------
# Replace missing Age values with the median age.
# ------------------------------------------------

median_age = df["Age"].median()

df["Age"] = df["Age"].fillna(median_age)

print("\nMedian Age Used :", median_age)

print("\nMissing Age Values After Cleaning")

print(df["Age"].isnull().sum())
````

### Interpretation

The missing values in the **Age** column have now been replaced using the median age.

After this preprocessing step:

- Every passenger has a valid age value.
- No missing values remain in the Age column.
- The dataset is now more suitable for machine learning.

Using the median helps preserve the overall distribution of the Age feature without being heavily influenced by unusually high or low ages.

> ## 💡 Machine Learning Note
>
> For numerical features:
>
> - Mean is suitable when the data is approximately symmetric and contains few outliers.
> - Median is preferred when the data is skewed or contains outliers.
>
> Since passenger ages can vary widely, the median is generally the safer choice.

### 🎓 Viva Question

**Q. Why did we fill missing Age values instead of deleting those rows?**

**Answer**

Removing all rows with missing Age values would reduce the amount of training data available to the model.

By replacing the missing values with the median, we retain all passenger records while providing a reasonable estimate for the missing ages. This helps preserve valuable information and improves the usefulness of the dataset.

# Step 10.3 — Handling Missing Values in the Embarked Column

The **Embarked** feature indicates the port from which each passenger boarded the Titanic.

This feature is categorical because it contains text values such as:

- C = Cherbourg
- Q = Queenstown
- S = Southampton

During our missing value analysis, we observed that only a very small number of entries are missing.

For categorical variables, the most common technique is to replace missing values with the **mode**, which is the most frequently occurring category.

Since only a few values are missing, replacing them with the mode will have a negligible effect on the overall distribution of the data.

````python
# ================================================================
# STEP 10.3 : HANDLE MISSING EMBARKED VALUES
# ================================================================

# Count missing values before cleaning.

print("=" * 60)
print("Missing Embarked Values Before Cleaning")
print("=" * 60)

print(df["Embarked"].isnull().sum())

# ------------------------------------------------
# Find the most frequent category (mode).
# ------------------------------------------------

most_common_port = df["Embarked"].mode()[0]

print("\nMost Frequent Embarkation Port :", most_common_port)

# ------------------------------------------------
# Replace missing values with the mode.
# ------------------------------------------------

df["Embarked"] = df["Embarked"].fillna(most_common_port)

print("\nMissing Embarked Values After Cleaning")

print(df["Embarked"].isnull().sum())
````

# Step 10.3A — Visualizing the Embarked Feature

After filling the missing values, it is useful to visualize the distribution of passengers across the three embarkation ports.

This visualization allows us to verify that replacing the missing values with the mode has not significantly changed the distribution of the categorical feature.

````python
# ================================================================
# VISUALIZE EMBARKATION PORTS
# ================================================================

plt.figure(figsize=(7,5))

sns.countplot(data=df, x="Embarked")

plt.title("Passenger Distribution by Embarkation Port")

plt.xlabel("Embarkation Port")

plt.ylabel("Number of Passengers")

plt.grid(axis="y", alpha=0.3)

plt.show()
````

### Interpretation

The bar chart shows the number of passengers boarding from each embarkation port.

We observe that:

- **Southampton (S)** has the highest number of passengers.

- **Queenstown (Q)** has the smallest number of passengers.

Since only two missing values were replaced with the most common category (Southampton), the overall distribution of the feature remains virtually unchanged.

This confirms that mode imputation was an appropriate preprocessing technique for this categorical feature.

> ## 💡 Machine Learning Note
>
> Numerical features are commonly filled using:
>
> - Mean
> - Median
>
> Categorical features are commonly filled using:
>
> - Mode (Most Frequent Category)
>
> because categorical values do not have numerical averages.

> ## 🚀 Professional Tip
>

### 🎓 Viva Question

**Q. Why can't we replace missing categorical values with the mean?**

**Answer**

The mean is a mathematical concept that applies only to numerical data.

Categorical features such as **Embarked** contain labels rather than numerical values, so calculating an average is not meaningful.

Instead, the most appropriate replacement is the **mode**, which represents the most frequently occurring category.

# EDA Section 10 — Family Size Analysis

The Titanic dataset includes two features related to family members:

- **SibSp**: Number of siblings and spouses aboard.
- **Parch**: Number of parents and children aboard.

Rather than analyzing these features separately, we can combine them into a new feature called **FamilySize**.

The formula is:

**FamilySize = SibSp + Parch + 1**

The additional **1** represents the passenger.

Creating meaningful new features from existing ones is known as **Feature Engineering**, an important step in many machine learning projects.

````python
# ================================================================
# CREATE FAMILY SIZE FEATURE
# ================================================================

df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

print("=" * 60)
print("First 10 Family Size Values")
print("=" * 60)

display(df[["SibSp", "Parch", "FamilySize"]].head(10))
````

````python
# ================================================================
# FAMILY SIZE DISTRIBUTION
# ================================================================

plt.figure(figsize=(9,5))

sns.countplot(
    data=df,
    x="FamilySize"
)

plt.title("Distribution of Family Size")

plt.xlabel("Family Size")

plt.ylabel("Number of Passengers")

plt.grid(axis="y", alpha=0.3)

plt.show()
````

````python
# ================================================================
# SURVIVAL BY FAMILY SIZE
# ================================================================

plt.figure(figsize=(10,5))

sns.barplot(
    data=df,
    x="FamilySize",
    y="Survived",
    estimator="mean",
    errorbar=None
)

plt.title("Survival Rate by Family Size")

plt.xlabel("Family Size")

plt.ylabel("Average Survival Rate")

plt.grid(axis="y", alpha=0.3)

plt.show()
````

### Interpretation

The FamilySize feature provides additional insight into passenger survival.

Observations may include:

- Passengers traveling completely alone often had different survival rates than those traveling with family.
- Very large families generally had lower survival rates.
- Small families may have experienced comparatively higher survival rates.

These patterns suggest that FamilySize captures information not directly visible in the original SibSp and Parch features.

> ## 💡 Machine Learning Note

> Feature Engineering is the process of creating new features from existing data.
>
> Well-designed features often improve model performance more than changing the machine learning algorithm itself.

> ## 🚀 Professional Tip

> Before creating new features, think about the underlying problem rather than combining variables randomly.

# Section 2 — Handling Missing Values in Age

The **Age** feature contains missing values.

Removing every row with a missing age would unnecessarily reduce the size of the dataset.

# Section 2 — Splitting the Dataset

# Section 3 — Feature Scaling

The features in the Titanic dataset have different numerical ranges.

For example:

- **Age** ranges roughly from 0 to 80.
- **Fare** ranges from 0 to over 500.
- **Pclass** ranges only from 1 to 3.

Machine learning algorithms that rely on distances, such as Support Vector Machines (SVM), are sensitive to these differences.

If the features are not scaled, variables with larger values can dominate the learning process.

To avoid this problem, we standardize the features using **StandardScaler**.

# What is StandardScaler?

StandardScaler transforms each feature so that:

- **Mean ≈ 0**
- **Standard Deviation ≈ 1**

The transformation is performed using the formula:

\[
z = \frac{x-\mu}{\sigma}
\]

Where:

- **x** = original value
- **μ** = mean of the training feature
- **σ** = standard deviation of the training feature

This process is called **standardization** or **Z-score normalization**.

````python
# ================================================================
# FEATURE SCALING USING STANDARDSCALER
# ================================================================

from sklearn.preprocessing import StandardScaler

# Create the scaler object.
scaler = StandardScaler()

# Learn scaling parameters from the training data only.
X_train_scaled = scaler.fit_transform(X_train)

# Apply the same transformation to the test data.
X_test_scaled = scaler.transform(X_test)

print("=" * 60)
print("Training Data Shape")
print("=" * 60)
print(X_train_scaled.shape)

print("\n")

print("=" * 60)
print("Testing Data Shape")
print("=" * 60)
print(X_test_scaled.shape)
````

````python
# ================================================================
# CONVERT SCALED ARRAYS TO DATAFRAMES
# ================================================================

X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X_test.columns,
    index=X_test.index
)

print("=" * 60)
print("First Five Rows of Scaled Training Data")
print("=" * 60)

display(X_train_scaled.head())
````

````python
# ================================================================
# VERIFY STANDARDIZATION
# ================================================================

summary = pd.DataFrame({
    "Mean": X_train_scaled.mean().round(2),
    "Standard Deviation": X_train_scaled.std().round(2)
})

display(summary)
````

````python
# ================================================================
# COMPARE FEATURE VALUES BEFORE AND AFTER SCALING
# ================================================================

comparison = pd.DataFrame({
    "Original Age": X_train["Age"].head(10),
    "Scaled Age": X_train_scaled["Age"].head(10),
    "Original Fare": X_train["Fare"].head(10),
    "Scaled Fare": X_train_scaled["Fare"].head(10)
})

display(comparison)
````

# Why Do We Fit Only on the Training Data?

Notice the following code:

```python
scaler.fit(X_train)
```

The scaler learns the **mean** and **standard deviation** from the training data only.

Later, we apply exactly the same transformation to the testing data:

```python
X_test_scaled = scaler.transform(X_test)
```

We never call:

```python
scaler.fit(X_test)
```

Doing so would allow information from the testing data to influence the preprocessing step, leading to **data leakage** and an unrealistic evaluation of the model.

### Interpretation

The feature scaling process has standardized all numerical features.

After scaling:

- Features have approximately zero mean.
- Features have approximately unit standard deviation.
- Large-valued features no longer dominate smaller-valued features.
- The dataset is now suitable for training a Support Vector Machine.

Proper scaling improves both the accuracy and stability of distance-based machine learning algorithms.

> ## 💡 Machine Learning Note
>
> Support Vector Machines, K-Nearest Neighbors, Logistic Regression, Neural Networks, and K-Means clustering generally perform better when features are standardized.
>
> Tree-based algorithms such as Decision Trees and Random Forests typically do not require feature scaling.

> ## 🚀 Professional Tip

>

### 🎓 Viva Question

**Q. Why is feature scaling important for Support Vector Machines?**

**Answer**

Support Vector Machines calculate distances and margins between data points. If one feature has much larger numerical values than others, it can dominate these calculations and reduce model performance. Feature scaling ensures that all features contribute more equally during training.

# Introduction to Support Vector Machine (SVM)

Support Vector Machine (SVM) is one of the most powerful supervised machine learning algorithms used for **classification** and **regression** problems.

In this project, we use SVM for **binary classification** because the target variable has only two possible outcomes:

- Survived (1)
- Did Not Survive (0)

The main objective of SVM is to find the **best decision boundary** that separates different classes while maximizing the distance (called the **margin**) between them.

A larger margin generally leads to better generalization on unseen data.

# How Does an SVM Work?

Imagine plotting passengers on a graph using their features.

The SVM algorithm searches for the **optimal hyperplane** that separates the two classes.

Among all possible separating lines (or hyperplanes), SVM selects the one with the **maximum margin**.

The data points that lie closest to this boundary are called **Support Vectors** because they determine the position of the hyperplane.

These support vectors are the most influential observations in the model.

# Choosing the Kernel Function

The **kernel** determines how SVM separates the data.

Some commonly used kernels include:

- Linear
- Polynomial
- Radial Basis Function (RBF)
- Sigmoid

For the Titanic dataset, we will use the **Radial Basis Function (RBF)** kernel.

The RBF kernel can model both linear and non-linear relationships, making it a strong default choice for many classification problems.

> ## 💡 Machine Learning Note
>
> The RBF kernel is often the default choice because it performs well on many real-world datasets without requiring extensive feature engineering.

### 🎓 Viva Question

**Q. Why did we choose the RBF kernel instead of a Linear kernel?**

**Answer**

The relationship between the passenger features and survival is unlikely to be perfectly linear. The RBF kernel can model more complex decision boundaries, making it a suitable choice for this classification problem.

# Training the Support Vector Machine

Training is the process during which the algorithm learns patterns from the training data.

The SVM algorithm analyzes the scaled training features and the corresponding survival labels to determine the optimal decision boundary.

Once training is complete, the model can classify previously unseen passengers.

````python
# ================================================================
# IMPORT SUPPORT VECTOR CLASSIFIER
# ================================================================

from sklearn.svm import SVC

# ================================================================
# CREATE THE MODEL
# ================================================================
svm_model = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale",
    probability=True,
    random_state=42
)
# ================================================================
# TRAIN THE MODEL
# ================================================================

svm_model.fit(
    X_train_scaled,
    y_train
)

print("=" * 60)
print("Support Vector Machine Model")
print("=" * 60)

print("Model training completed successfully.")
````

# Understanding the Model Parameters

The Support Vector Machine was created using the following parameters:

### kernel = "rbf"

Uses the Radial Basis Function kernel to learn non-linear decision boundaries.

### C = 1.0

Controls the balance between maximizing the margin and minimizing classification errors.

- Smaller C → Simpler model with a wider margin.
- Larger C → More complex model with fewer training errors.

### gamma = "scale"

Automatically determines an appropriate influence radius for each training example based on the dataset.

### random_state = 42

Ensures reproducibility where applicable in the workflow.

### Interpretation

The Support Vector Machine has now learned from the training data.

At this stage, the model has **not yet been evaluated**. It has only identified patterns within the training dataset.

# Accuracy Score

Accuracy is the proportion of correctly classified observations among all predictions.

It is calculated using the formula:

\[
Accuracy = \frac{Correct\ Predictions}{Total\ Predictions}
\]

Accuracy is easy to understand and is often the first metric reported for classification models.

- True Positive (TP)

# ROC Curve (Receiver Operating Characteristic)

The ROC Curve is a graphical method for evaluating the performance of a binary classification model.

Instead of measuring only how many predictions are correct, the ROC Curve examines the model's performance across different classification thresholds.

The curve plots:

- **True Positive Rate (TPR)** on the Y-axis
- **False Positive Rate (FPR)** on the X-axis

A model with better discrimination ability produces a curve that is closer to the upper-left corner of the graph.

# What is AUC?

The Area Under the ROC Curve (AUC) summarizes the ROC Curve using a single numerical value.

The AUC ranges from 0 to 1.

General interpretation:

- **1.00** → Perfect classifier
- **0.90 – 0.99** → Excellent
- **0.80 – 0.89** → Very Good
- **0.70 – 0.79** → Acceptable
- **0.60 – 0.69** → Poor
- **0.50** → No better than random guessing

Higher AUC values indicate better discrimination between the two classes.

# Introduction

Earlier we trained an SVM using the default parameters.

Although the model performed well, the default settings are not necessarily optimal for every dataset.

Machine Learning models often achieve better performance after carefully tuning their **hyperparameters**.

Instead of selecting values by trial and error, we will use **GridSearchCV**, which automatically evaluates multiple parameter combinations using cross-validation.

This systematic approach helps identify the parameter values that produce the best model performance.

# What are Hyperparameters?

Hyperparameters are settings that are chosen **before** the learning process begins.

Unlike model parameters, which are learned automatically during training, hyperparameters control how the algorithm learns.

For Support Vector Machines, some important hyperparameters include:

- Kernel
- C
- Gamma

Selecting appropriate values can significantly improve classification performance.

# Hyperparameters Used in This Project

### Kernel

Determines the type of decision boundary.

We will compare:

- Linear
- RBF

---

### C

Controls the trade-off between:

- wider margin
- fewer classification errors

Small C

→ simpler model

Large C

→ more flexible model

---

### Gamma

Determines how much influence each training example has.

Small Gamma

→ smoother decision boundary

Large Gamma

→ more complex decision boundary

````python
# ================================================================
# IMPORT GRIDSEARCHCV
# ================================================================

from sklearn.model_selection import GridSearchCV
````

````python
# ================================================================
# PARAMETER GRID
# ================================================================

param_grid = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto"]
}

param_grid
````

# Why These Values?

Rather than searching every possible value, we define a reasonable search space.

The parameter grid contains several combinations of:

- C values
- Gamma values
- Kernel types

GridSearchCV will train and evaluate an SVM model for **every combination** and automatically select the best-performing configuration.

````python
# ================================================================
# GRID SEARCH
# ================================================================

grid_search = GridSearchCV(

    estimator=SVC(),

    param_grid=param_grid,

    cv=5,

    scoring="f1_weighted",

    n_jobs=1,

    verbose=2
)

grid_search.fit(
    X_train_scaled,
    y_train
)
````

# Cross Validation

Instead of evaluating the model on a single train-validation split, GridSearchCV uses **5-fold Cross Validation**.

The training data is divided into five equal parts.

The model is trained five times.

Each fold is used once as the validation set.

The average validation accuracy is used to compare parameter combinations.

Cross-validation generally produces a more reliable estimate of model performance.

````python
# ================================================================
# BEST PARAMETERS
# ================================================================

print("="*60)
print("Best Parameters")
print("="*60)

print(grid_search.best_params_)
````

````python
# ================================================================
# BEST CROSS VALIDATION SCORE
# ================================================================

print("="*60)
print("Best Cross Validation Accuracy")
print("="*60)

print(grid_search.best_score_)
````

````python
# ================================================================
# BEST MODEL
# ================================================================

from sklearn.svm import SVC

best_model = SVC(
    C=grid_search.best_params_["C"],
    kernel=grid_search.best_params_["kernel"],
    gamma=grid_search.best_params_["gamma"],
    probability=True,
    random_state=42
)

best_model.fit(
    X_train_scaled,
    y_train
)
````

````python
# ================================================================
# PREDICTIONS USING OPTIMIZED MODEL
# ================================================================

best_pred = best_model.predict(
    X_test_scaled
)
````

````python
# ================================================================
# ACCURACY OF OPTIMIZED MODEL
# ================================================================

best_accuracy = accuracy_score(
    y_test,
    best_pred
)

print(best_accuracy)
````

````python
# ================================================================
# PERFORMANCE COMPARISON
# ================================================================

comparison = pd.DataFrame({

    "Model":[
        "Default SVM",
        "Optimized SVM"
    ],

    "Accuracy":[
        accuracy,
        best_accuracy
    ]
})

comparison["Accuracy (%)"] = (
    comparison["Accuracy"]*100
).round(2)

display(comparison)
````

````python
# ================================================================
# BAR CHART
# ================================================================

plt.figure(figsize=(7,5))

plt.bar(
    comparison["Model"],
    comparison["Accuracy"]
)

plt.ylim(0,1)

plt.ylabel("Accuracy")

plt.title("Default vs Optimized SVM")

plt.grid(axis="y",alpha=0.3)

plt.show()
````

### Interpretation

GridSearchCV evaluated multiple parameter combinations using five-fold cross-validation.

The optimized model achieved the highest validation accuracy among all tested combinations.

Comparing the optimized model with the default SVM allows us to determine whether hyperparameter tuning produced a meaningful improvement.

> ## 💡 Machine Learning Note
>
> Hyperparameter tuning aims to improve the model's ability to generalize to unseen data. Better performance is not guaranteed, but a systematic search helps identify stronger configurations than relying solely on default settings.

> ## 🚀 Professional Tip
>
> Avoid choosing hyperparameters by repeatedly testing on the test set. Perform tuning with cross-validation on the training data, then evaluate the final optimized model only once on the untouched test set.

# Why Save a Machine Learning Model?

Training a machine learning model can take time and computational resources.

Once a model has been trained, there is no need to retrain it every time we want to make a prediction.

Instead, we can save the trained model to disk and load it whenever it is needed.

This approach is commonly used in production systems, web applications, and machine learning deployment pipelines.

# Why Use Joblib?

Python provides several methods for saving objects.

For machine learning models, **Joblib** is generally preferred because it is optimized for storing objects containing large NumPy arrays.

In this project, we will save:

- Optimized SVM model
- StandardScaler
- Sex Label Encoder
- Embarked Label Encoder

Saving all preprocessing objects ensures that future predictions use exactly the same transformations that were applied during training.

````python
# ================================================================
# IMPORT JOBLIB
# ================================================================

import joblib
````

# Loading Saved Objects

A saved machine learning model can be restored using `joblib.load()`.

Once loaded, the model behaves exactly like the original trained model.

There is no need to retrain the algorithm.

# Predicting New Passenger Survival

Training and evaluating a machine learning model are important steps, but the ultimate goal is to use the trained model to make predictions for new, unseen data.

In this chapter, we will build a reusable prediction function that:

- Accepts passenger information
- Applies the same preprocessing used during training
- Scales numerical features
- Uses the trained SVM model to predict survival
- Returns both the predicted class and the prediction probability

This modular design makes the model easy to reuse in web applications, APIs, and deployment platforms such as Gradio and Hugging Face Spaces.

````python
# ================================================================
# PASSENGER SURVIVAL PREDICTION FUNCTION
# ================================================================

def predict_survival(
    pclass,
    sex,
    age,
    sibsp,
    parch,
    fare,
    embarked
):
    """
    Predict Titanic passenger survival.

    Parameters
    ----------
    pclass : int
    sex : str
    age : float
    sibsp : int
    parch : int
    fare : float
    embarked : str

    Returns
    -------
    prediction
    probability
    """

    # ------------------------------------------------------------
    # Calculate engineered feature
    # ------------------------------------------------------------

    family_size = sibsp + parch + 1

    # ------------------------------------------------------------
    # Encode categorical variables
    # ------------------------------------------------------------

    sex_encoded = loaded_sex_encoder.transform([sex])[0]

    embarked_encoded = loaded_embarked_encoder.transform(
        [embarked]
    )[0]

    # ------------------------------------------------------------
    # Create DataFrame
    # ------------------------------------------------------------

    passenger = pd.DataFrame({

        "Pclass":[pclass],

        "Sex":[sex_encoded],

        "Age":[age],

        "SibSp":[sibsp],

        "Parch":[parch],

        "Fare":[fare],

        "Embarked":[embarked_encoded],

        "FamilySize":[family_size]

    })

    # ------------------------------------------------------------
    # Scale features
    # ------------------------------------------------------------

    passenger_scaled = loaded_scaler.transform(passenger)

    # ------------------------------------------------------------
    # Prediction
    # ------------------------------------------------------------

    prediction = loaded_model.predict(
        passenger_scaled
    )[0]

    probability = loaded_model.predict_proba(
        passenger_scaled
    )[0]

    return prediction, probability
````

The following example represents a hypothetical Titanic passenger.

We will use the trained machine learning model to predict whether this passenger would likely survive.

````python
# ================================================================
# EXAMPLE PASSENGER
# ================================================================

prediction, probability = predict_survival(

    pclass=1,

    sex="female",

    age=28,

    sibsp=0,

    parch=0,

    fare=85,

    embarked="S"

)
````
